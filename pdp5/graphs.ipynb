```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
num_of_threads = [2, 4, 8, 16, 20]

# Sequential real times
seq_40_8_15 = 469.771
seq_35_10_18 = 419.177
seq_45_5_16 = 336.617

# Task real times (for 2, 4, 8, 16, 20 threads)
task_40_8_15_1 = [113.322, 48.537, 26.981, 20.594, 20.384]
task_35_10_18_2 = [83.16, 35.456, 25.109, 15.427, 20.824]
task_45_5_16_3 = [41.23, 34.213, 9.456, 7.394, 4.626]

# Task acceleration
a_task_40_8_15_1 = [seq_40_8_15 / val for val in task_40_8_15_1]
a_task_35_10_18_2 = [seq_35_10_18 / val for val in task_35_10_18_2]
a_task_45_5_16_3 = [seq_45_5_16 / val for val in task_45_5_16_3]

# Task efficiency
e_task_40_8_15_1 = [val / num for val, num in zip(a_task_40_8_15_1, num_of_threads)]
e_task_35_10_18_2 = [val / num for val, num in zip(a_task_35_10_18_2, num_of_threads)]
e_task_45_5_16_3 = [val / num for val, num in zip(a_task_45_5_16_3, num_of_threads)]

# Data real times (for 2, 4, 8, 16, 20 threads)
data_40_8_15_1 = [169.248, 81.64, 53.83, 51.615, 52.027]
data_35_10_18_2 = [91.39, 46.001, 27.065, 24.183, 25.404]
data_45_5_16_3 = [70.951, 32.233, 18.005, 14.778, 13.801]

# Data acceleration
a_data_40_8_15_1 = [seq_40_8_15 / val for val in data_40_8_15_1]
a_data_35_10_18_2 = [seq_35_10_18 / val for val in data_35_10_18_2]
a_data_45_5_16_3 = [seq_45_5_16 / val for val in data_45_5_16_3]

# Data efficiency
e_data_40_8_15_1 = [val / num for val, num in zip(a_data_40_8_15_1, num_of_threads)]
e_data_35_10_18_2 = [val / num for val, num in zip(a_data_35_10_18_2, num_of_threads)]
e_data_45_5_16_3 = [val / num for val, num in zip(a_data_45_5_16_3, num_of_threads)]

# MPI real times (for 4 processes and 2, 4, 8, 16, 20 threads)
mpi_40_8_15_1 = [122.651, 61.119, 35.52, 40.605, 41.618]
mpi_35_10_18_2 = [102.979, 53.837, 32.64, 31.665, 31.165]
mpi_45_5_16_3 = [94.702, 48.054, 31.436, 30.293, 30.899]

# MPI acceleration
a_mpi_40_8_15_1 = [seq_40_8_15 / val for val in mpi_40_8_15_1]
a_mpi_35_10_18_2 = [seq_35_10_18 / val for val in mpi_35_10_18_2]
a_mpi_45_5_16_3 = [seq_45_5_16 / val for val in mpi_45_5_16_3]

# MPI efficiency
e_mpi_40_8_15_1 = [val / (1 + 3 * num) for val, num in zip(a_mpi_40_8_15_1, num_of_threads)]
e_mpi_35_10_18_2 = [val / (1 + 3 * num) for val, num in zip(a_mpi_35_10_18_2, num_of_threads)]
e_mpi_45_5_16_3 = [val / (1 + 3 * num) for val, num in zip(a_mpi_45_5_16_3, num_of_threads)]
```

```python
plt.subplots(figsize = (12, 6))
plt.xticks(range(0, 22, 2))
plt.xlabel("Počet vláken", fontweight = "bold", fontsize = 15)
plt.ylabel("Reálný čas (s)", fontweight = "bold", fontsize = 15)
plt.plot(num_of_threads, task_40_8_15_1, marker = "o", linestyle = "solid", color = "red", label = "Funkční paralelismus")
plt.plot(num_of_threads, data_40_8_15_1, marker = "o", linestyle = "solid", color = "green", label = "Datový paralelismus")
plt.plot(num_of_threads, mpi_40_8_15_1, marker = "o", linestyle = "solid", color = "blue", label = "MPI (4 procesy)")
plt.grid()
plt.legend()
plt.show()

plt.subplots(figsize = (12, 6))
plt.xticks(range(0, 22, 2))
plt.xlabel("Počet vláken", fontweight = "bold", fontsize = 15)
plt.ylabel("Zrychlení", fontweight = "bold", fontsize = 15)
plt.plot(num_of_threads, a_task_40_8_15_1, marker = "o", linestyle = "solid", color = "red", label = "Funkční paralelismus")
plt.plot(num_of_threads, a_data_40_8_15_1, marker = "o", linestyle = "solid", color = "green", label = "Datový paralelismus")
plt.plot(num_of_threads, a_mpi_40_8_15_1, marker = "o", linestyle = "solid", color = "blue", label = "MPI (4 procesy)")
plt.plot(np.linspace(2, 20, 100), np.linspace(2, 20, 100), linestyle = "dotted", color = "black", label = "Lineární funkce")
plt.grid()
plt.legend()
plt.show()

plt.subplots(figsize = (12, 6))
plt.xticks(range(0, 22, 2))
plt.xlabel("Počet vláken", fontweight = "bold", fontsize = 15)
plt.ylabel("Efektivnost", fontweight = "bold", fontsize = 15)
plt.plot(num_of_threads, e_task_40_8_15_1, marker = "o", linestyle = "solid", color = "red", label = "Funkční paralelismus")
plt.plot(num_of_threads, e_data_40_8_15_1, marker = "o", linestyle = "solid", color = "green", label = "Datový paralelismus")
plt.plot(num_of_threads, e_mpi_40_8_15_1, marker = "o", linestyle = "solid", color = "blue", label = "MPI (4 procesy)")
plt.axhline(y = 1, linestyle = "dotted", color = "black", label = "Horní mez")
plt.grid()
plt.legend()
plt.show()
```
